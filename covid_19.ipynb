```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
```

```python
confirmed11=pd.read_csv('time_series_covid19_confirmed_global.csv')
```

```python
confirmed11
```

```python
deaths11=pd.read_csv('time_series_covid19_deaths_global.csv')
deaths11
```

```python
recovered11=pd.read_csv('time_series_covid19_recovered_global.csv')
recovered11
```

```python
confirmed11=confirmed11.drop(['Province/State','Lat','Long'],axis=1).groupby('Country/Region').sum()
confirmed11
```

```python
deaths11=deaths11.drop(['Province/State','Lat','Long'],axis=1).groupby('Country/Region').sum()
deaths11
```

```python
recovered11=recovered11.drop(['Province/State','Lat','Long'],axis=1).groupby('Country/Region').sum()
recovered11
```

```python
confirmed=confirmed11.T
confirmed
```

```python
deaths=deaths11.T
deaths
```

```python
recovered=recovered11.T
recovered
```

```python
confirmed.index.name='Date'
confirmed=confirmed.reset_index()
confirmed
```

```python
recovered.index.name='Date'
recovered=recovered.reset_index()
recovered
```

```python
deaths.index.name='Date'
deaths=deaths.reset_index()
```

```python
confirmed_melt=confirmed.melt(id_vars='Date').copy()
confirmed_melt
```

```python
recovered_melt=recovered.melt(id_vars='Date').copy()
recovered_melt
```

```python
deaths_melt=deaths.melt(id_vars='Date').copy()
deaths_melt
```

```python
confirmed_melt.rename(columns={'value':'confirm'},inplace=True)
confirmed_melt
```

```python
deaths_melt.rename(columns={'value':'deaths'},inplace=True)
deaths_melt
```

```python
recovered_melt.rename(columns={'value':'recover'},inplace=True)
recovered_melt
```

```python
data11=pd.merge(confirmed_melt,deaths_melt,on=['Date','Country/Region'] ,how='right')
data11
```

```python
data12=pd.merge(data11,recovered_melt,on=['Date','Country/Region'] ,how='right')
data12
```
